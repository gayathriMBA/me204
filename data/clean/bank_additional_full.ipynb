## AGE GROUP

```python
df['age_group']= df['age'].apply(lambda x:'-30'if x <=30 
                                 else '31-40'if x <=40
                                 else'40-50' if x <=50
                                 else'50-60' if x <=60
                                 else '60+'
                                )
```

### High Conversion Age Group:

The 60+ age group shows the highest conversion rate at 45.49%. This indicates that older individuals are the most likely to convert.
Moderate Conversion Age Groups:

The 50-60 age group has a conversion rate of 10.65%, which is moderate.
The -30 age group (younger individuals) has a conversion rate of 15.22%, which is slightly higher than the 50-60 range.
Low Conversion Age Groups:

The 31-40 age group has a relatively low conversion rate of 9.75%.
The 40-50 age group shows the lowest conversion rate at 8.17%, making them the least likely to convert.

# pivot table

why we not use the count because the count cannot take a NA so we use len

```python
status_df=pd.pivot_table(df, values='y',index='marital',columns='conversion',aggfunc=len)
status_df
```

```python
status_df.columns=['non_conversion','conversion']
```

```python
status_df
```

Married individuals are more likely to convert, while singles and divorced individuals have higher non-conversion probabilities. Conversion campaigns should target married demographics for effectiveness and improve engagement strategies for singles and divorced individuals.

```python
status_df.plot(kind='pie',
      figsize=(15,7),
      startangle=90,
      subplots=True,
      autopct=lambda x: '%0.1f%%'%x
)
plt.show
```

# EDUCATION

```python
conversions_by_education_df =pd.pivot_table(df, values='y',index='education',columns='conversion',aggfunc=len)
conversions_by_education_df
```

```python
conversions_by_education_df.columns =['non_conversion','conversion']
conversions_by_education_df
```

higher education peoples are more likely to convert 

```python
conversions_by_education_df.plot(kind='pie',
      figsize=(15,7),
      startangle=90,
      subplots=True,
      autopct=lambda x: '%0.1f%%'%x
)
plt.show()
```
