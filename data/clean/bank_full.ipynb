# Encoding

```python
encode_df=pd.get_dummies(df,columns=['job','marital'],dtype=int)
full_df=pd.concat([df,encode_df],axis=1)
```

```python
full_df=full_df.loc[:,~full_df.columns.duplicated()]
```

```python
df['housing']= df['housing'].apply(lambda x:1 if x=='yes' else 0)
df['loan']= df['loan'].apply(lambda x:1 if x=='yes' else 0)
```

```python
list(full_df.columns)
```

# decision tree

```python
#! pip install scikit-learn

from sklearn import tree
```

```python
response_var = 'conversion'
features = [
    'age','balance','campaign','previous','housing',
 'loan','job_management',
 'job_retired',
 'job_self-employed',
 'job_services',
 'job_student',
 'job_technician',
 'job_unemployed',
 'job_unknown',
 'marital_divorced',
 'marital_married',
 'marital_single'
]
```

```python
dt_model = tree.DecisionTreeClassifier(
     max_depth=4  # depth code is for the 
)
```

```python
dt_model 
```

```python
dt_model.fit(full_df[features],df[response_var])
```

```python
dt_model.classes_
```

## we identitfy which one is highly important in the variable so we use feature_importances  ( highly significant is age nama  features la variable store pannirukura order ladha varum yedha output pota adha variable )

```python
dt_model.feature_importances_


```

```python
import matplotlib.pyplot as plt

#convert class names to strings
class_names = [str(label) for label in dt_model.classes_]
```

##### previous(campaign)<=1 more than one campaign attend pannavaga rate is high , campaign less than 1 errukum bodhu less conversion rate 
##### avoid blue-collor job , management and student job are more conversion , younger peoples based on job, marital  conversion rate is high.
##### Target younger individuals.  Prioritize those who have participated in multiple campaigns.Focus on specific job roles like management and students while avoiding blue-collar workers.

```python
plt.figure(figsize=(15,10))
tree.plot_tree(dt_model,feature_names=features,class_names=class_names,filled=True)
plt.show()
```
